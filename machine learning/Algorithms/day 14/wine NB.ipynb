Importing essential libraries
such as numpy, pandas, matplotlib

also load the data set **wine**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

loading data set

```python
wine = load_wine()
```

Data set Description

```python
wine.DESCR
```

Data Feature name

```python
wine.feature_names
```

converting to a pandas DataFrame

```python
df = pd.DataFrame(wine.data,columns=wine.feature_names)
```

Data Frame

```python
df
```

target value adding to Data Frame

```python
df['target'] = wine.target
```

```python
df
```

Separate the data to give train test model

```python
x = df.drop(['alcohol',
 'malic_acid',
 'ash',
 'alcalinity_of_ash',
 'magnesium','nonflavanoid_phenols','target'],axis='columns')
y = df.target
```

import the model selection library

```python
from sklearn.model_selection import train_test_split
```

split the train - test data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

our Model

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB
```

GuassianNB

```python
model = GaussianNB()
```

fit model

```python
model.fit(x_train,y_train)
```

model Score

```python
model.score(x_test,y_test)
```

MultinomianlNB

```python
model1 = MultinomialNB()
```

fit the model

```python
model1.fit(x_train,y_train)
```

model score

```python
model1.score(x_test,y_test)
```
